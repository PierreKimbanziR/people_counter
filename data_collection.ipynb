```python
import fiftyone as fo
import fiftyone.zoo as foz
dataset = foz.load_zoo_dataset("open-images-v6", split="train", classes=["Man", "Boy", "Girl", "Woman"], label_types = "detections",max_samples = 8000)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/root/fiftyone/open-images-v6/train/labels/detections.csv')
```

```python
df.head()
```

```python
import os 
from pathlib import Path
path = "/root/fiftyone/open-images-v6/train/data"
image_ids = [Path(image).stem for image in os.listdir(path)]


```

```python
len(image_ids)
```

```python
labels = {"/m/01bl7v" : "Boy", "/m/04yx4" : "Man", "/m/03bt1vf" : "Woman", "/m/05r655" : "Girl" }
label_names = labels.keys()
```

```python
def create_df_with_relevant_images():
  new_df = df.loc[df["ImageID"].isin(image_ids)]
  new_df = new_df.loc[new_df["LabelName"].isin(label_names)]
  return new_df
```

```python
new_df = create_df_with_relevant_images()
```

```python
def create_text_file():
  labels = {"/m/01bl7v" : "Boy", "/m/04yx4" : "Man", "/m/03bt1vf" : "Woman", "/m/05r655" : "Girl" }

  for row in new_df.iterrows():
    image_id = row[1]["ImageID"]
    label = labels[row[1]["LabelName"]]
    x_min = row[1]["XMin"]
    x_max = row[1]["XMax"]
    y_min = row[1]["YMin"]
    y_max = row[1]["YMax"]
    with open(f"/content/drive/MyDrive/Peoples/train/{image_id}.txt", "a") as text_file:
      text_file.write("\n")
      text_file.write(f"{label},{x_min},{x_max},{y_min},{y_max}")
    


```

```python
create_text_file()
```
