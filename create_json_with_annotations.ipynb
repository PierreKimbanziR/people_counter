# Create a Jsonfile with every labels and it respective bounding boxes
### How ?  

*   Iterate on every text file and retrieve the label and the bounding box corresponding to the label.

*   Store those values in a dictionary


 


## Import libraries

```python
import json
import os 
from pathlib import Path

```

```python
new_path = "/content/drive/MyDrive/Peoples/train"

def create_json():
  full_dict ={}
  for file in os.listdir(new_path):
    print(file)
    if file.endswith('.txt'):
      
      file_dict = {}
      file_name = Path(new_path+'/'+file).stem
      open_file = open(new_path+'/'+file)
      lines = [line for line in open_file.readlines() if line.strip()]
      for index, line in enumerate(lines):
        stripped_line = line.rstrip()
        list_line = stripped_line.split(",")
        line_dict = {}
        line_dict["bounding_box"] = list_line[1:]
        line_dict["label"] = list_line[0]
        file_dict[index] = line_dict
      open_file.close()
      full_dict[file_name] = file_dict
      
  with open('/content/drive/MyDrive/Peoples/train/labels.json', "a+") as js_file:
    json.dump(full_dict, js_file)
```

```python
create_json()
```
